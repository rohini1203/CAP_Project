```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
%matplotlib inline
```

### Download the dataset

```python
df= pd.read_csv('https://www.doogal.co.uk/UKPostcodesCSV.ashx?area=London')
df.head()
```

```python
df.shape
```

### Not all these features will be required, we will only get 8 of the 44 features

```python

df1=df[['Postcode','Latitude','Longitude','County','District','Ward','Nearest station','Distance to station']]
df1.head()
```

```python
df1.shape
```

### Now we shall select the rows where the county is 'City of London since it is our target area

```python
df2=df1[df1['County']=='City of London']
df2.head()
```

```python
df2.shape
```

### Get rows where with Postcode starting with 'E1"

```python
df3=df2[df2['Postcode'].str.contains('E1')].reset_index()
df3.head()
```

```python
df3.shape
```

```python
df4=df3[['Postcode','Latitude','Longitude','District','Ward','Nearest station','Distance to station']]
df4.head()
```
